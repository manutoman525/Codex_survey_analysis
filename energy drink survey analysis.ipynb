## Data Collection

```python
import pandas as pd

cities = pd.read_csv("E:\\codebasics\\energy drink data\\dim_cities.csv")
respondents = pd.read_csv('E:\\codebasics\\energy drink data\\dim_repondents.csv')
survey = pd.read_csv("E:\\codebasics\\energy drink data\\fact_survey_responses.csv")
```

## Demographic insights

### 1. Who prefers energy drink more? (male/female/non-binary?)

```python
pre_drnk_gen = df['Gender'].value_counts()
pre_drnk_gen
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10,5))
sns.set_style("whitegrid")
cls = sns.color_palette("viridis")
plt.pie(pre_drnk_gen , labels = pre_drnk_gen.index ,autopct = "%1.1f%%", shadow = True ,colors = cls)
plt.title("Consumers")

plt.show()
```

### 2. Which age group prefers energy drinks more?

```python
age_drnk = df['Age'].value_counts()
age_drnk
```

```python
bar = sns.barplot(x = age_drnk.index , y = age_drnk.values , palette = "mako" )
bar.bar_label(bar.containers[0])
bar.grid(False)
plt.ylabel("Count of consumers")
plt.title("Consumers according to Age")
plt.show()
```

### 3. which type of marketing reaches more to youth (15-30)

```python
plt.figure(figsize = (7,6))
bar= sns.barplot(x = high_markt.index , y = high_markt.values ,palette = "crest" )
bar.grid(False)
bar.bar_label(bar.containers[0])
plt.ylabel("Number of feedback")
plt.title("Marketing Channels  that reaches Youth")
plt.show()


```
